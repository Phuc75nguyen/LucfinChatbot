```python
import pandas as pd

# 1. Load d·ªØ li·ªáu th√¥ (gi·∫£ l·∫≠p b∆∞·ªõc Data Acquisition)
# ƒê·∫£m b·∫£o file foods.csv ƒëang n·∫±m c√πng th∆∞ m·ª•c
df = pd.read_csv('foods.csv')

# Ch·ªçn m·ªôt d√≤ng m·∫´u ƒë·ªÉ demo (V√≠ d·ª• d√≤ng ƒë·∫ßu ti√™n - M·ª±c ·ªëng h·∫•p)
sample_idx = 0
raw_row = df.iloc[sample_idx]

# ==============================================================================
# 2. ƒê·ªäNH NGHƒ®A THU·∫¨T TO√ÅN CHUY·ªÇN ƒê·ªîI (TRANSFORMATION ALGORITHM)
# ==============================================================================
def semantic_engineering(row):
    """
    H√†m th·ª±c hi·ªán 2 nhi·ªám v·ª•:
    1. Contextual Serialization: N·ªëi c√°c tr∆∞·ªùng r·ªùi r·∫°c th√†nh vƒÉn b·∫£n ng·ªØ nghƒ©a.
    2. Metadata Decoupling: T√°ch s·ªë li·ªáu v√† link ·∫£nh ra kh·ªèi vƒÉn b·∫£n embed.
    """
    # A. CHU·ªñI H√ìA NG·ªÆ C·∫¢NH (D√πng cho Vector Search)
    # K·ªπ thu·∫≠t: Explicit Labeling (G√°n nh√£n t∆∞·ªùng minh) gi√∫p AI hi·ªÉu c·∫•u tr√∫c
    contextual_text = (
        f"M√≥n ƒÉn: {row['dish_name']}\n"
        f"M√¥ t·∫£: {row['description']}\n"
        f"Th√†nh ph·∫ßn nguy√™n li·ªáu: {row['ingredients']}\n"
        f"C√°ch ch·∫ø bi·∫øn: {row['cooking_method']}"
    )
    
    # B. T√ÅCH BI·ªÜT METADATA (D√πng cho Filtering & UI)
    metadata = {
        "calories": int(row['calories']) if pd.notna(row['calories']) else 0,
        "protein": int(row['protein']) if pd.notna(row['protein']) else 0,
        "fat": int(row['fat']) if pd.notna(row['fat']) else 0,
        "image_link": str(row['image_link'])
    }
    
    return contextual_text, metadata

# ==============================================================================
# 3. TH·ª∞C THI V√Ä HI·ªÇN TH·ªä K·∫æT QU·∫¢ (VISUALIZATION)
# ==============================================================================
# √Åp d·ª•ng thu·∫≠t to√°n
sem_text, meta_struct = semantic_engineering(raw_row)

print("="*80)
print("DEMONSTRATION OF SEMANTIC DATA ENGINEERING PROCESS")
print("="*80)

print("\nüìç B∆Ø·ªöC 1: INPUT - RAW TABULAR DATA (D·ªØ li·ªáu th√¥ d·∫°ng b·∫£ng)")
print("-" * 60)
print(f"| Column: dish_name      | Value: {raw_row['dish_name']}")
print(f"| Column: ingredients    | Value: {str(raw_row['ingredients'])[:50]}...") 
print(f"| Column: cooking_method | Value: {str(raw_row['cooking_method'])[:50]}...")
print(f"| Column: calories       | Value: {raw_row['calories']}")
print(f"| Column: image_link     | Value: {raw_row['image_link']}")

print("\n" + "‚¨á" * 30 + " (Transformation Pipeline) " + "‚¨á" * 30)

print("\nüìç B∆Ø·ªöC 2: OUTPUT - SEMANTIC VECTOR NODE (Node ng·ªØ nghƒ©a ƒë√£ x·ª≠ l√Ω)")
print("-" * 60)
print("üÖ∞Ô∏è [Semantic Content] - D·ªØ li·ªáu d√πng ƒë·ªÉ Embedding (Vector h√≥a):")
print(f"'''\n{sem_text}\n'''")

print("\nüÖ±Ô∏è [Structured Metadata] - D·ªØ li·ªáu s·ªë h·ªçc & hi·ªÉn th·ªã (Kh√¥ng Embed):")
print(meta_struct)
print("="*80)
```
